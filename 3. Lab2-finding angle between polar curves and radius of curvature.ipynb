finding angle between polar curves

1)r=a(theta)& r=a/theta

```python
from sympy import *
r,theta,a=symbols('r, theta, a')
r1=a*theta;
r2=a/(theta);
dr1=diff(r1, theta);
dr2=diff(r2, theta);
t1=dr1/r1
t2=dr2/r2
q=solve(r1-r2, theta)
print ('q=',(q))
w1=t1.subs({theta:(q[0])})
print('w1=',(w1))
w2=t2.subs({theta:(q[0])})
print('w2 =' ,(w2))
y1=acot(w1)
print('y1=',(y1))
y2=acot(w2)
print('y2=' ,(y2))
w=abs(y1-y2)
print('w=',(w))
print('angle between two polar curves is', w)
```

2b) finding radius of curvature in cartesian, parametricform

1) find the radius of curvature sqrt(x) + sqrt(y)=4 at (4,4)

```python
from sympy import *
x,y=symbols('x y')
y=(4-sqrt(x))**2;
dy1=diff(y,x)
dy2=diff(dy1,x)
rho=simplify(1+dy1**2)**1.5/dy2;
rho1=rho.subs(x,4);
rho2=rho1.subs(y,4);
print('dy1=', (dy1))
print('dy2=', (dy2))
print('rho=', (rho))
print('radius of curvature at x=4 and y=4 is %0.3f'%(rho2))
```

2) find the radius of curvature of y^2=[4a^2(2a-x)]/x at (2a,0)

```python
from sympy import *
x,y,a=symbols('x y a')
x=(8*a**3/(y**2+4*a**2));
dx1=diff(x,y)
dx2=diff(dx1,y)
rho=simplify(1+dx1**2)**1.5/dx2;
rho1=rho.subs(x,2*a);
rho2=rho1.subs(y,0);
rho3=rho2.subs(a,2);
print('dx1=', (dx1))
print('dx2=', (dx2))
print('rho=', (rho))
print('radius of curvature at x=2a and y=0 is', simplify(rho3))
```

parametric form: 1) x=a[theta+sin(theta)] & y=b[1-cos(theta)] at theta=pi/2 & a=b=1

```python
from sympy import *
x,y,theta,a,b=symbols('x y theta a b')
x=a*(theta+sin(theta));
y=b*(1-cos(theta));
dx1=diff(x,theta)
dx2=diff(dx1,theta)
dy1=diff(y,theta)
dy2=diff(dy1,theta)
rho=((dx1**2+dy1**2)**1.5)/(dx1*dy2-dy1*dx2);
rho1=rho.subs(theta,pi/2);
rho2=rho1.subs(a,1);
rho3=rho2.subs(b,1);
print('dx1=', (dx1))
print('dx2=', (dx2))
print('dy1=', (dy1))
print('dy2=', (dy2))
print('rho=', (rho))
print('radius of curvature at x=pi/2 and a,b=1 is %0.3f '%(rho3))
```

polar form: 1) r=4(1+cos(theta)) at theta=pi/2

```python
from sympy import*
r,theta=symbols('r theta')
r=4*(1+cos(theta));
dr1=diff(r,theta)
dr2=diff(dr1,theta)
rho=(r**2+(dr1)**2)**1.5/(r**2+2*(dr1)**2-r*dr2);
rho1=rho.subs(theta,(pi/2));
print('dr1=',(dr1))
print('dr2=',(dr2))
print('rho=',(rho))
print('radius of curvature is %0.3f',(rho1));
```

2) r=a sin(n theta) at theta=pi/2,n=1,a=3

```python
from sympy import*
r,a,n,theta=symbols('r a n theta')
r=a*sin(a*theta);
dr1=diff(r,theta)
dr2=diff(dr1,theta)
rho=(r**2+(dr1)**2)**1.5/(r**2+2*(dr1)**2-r*dr2);
rho1=rho.subs(theta,(pi/2));
rho2=rho1.subs(n,1);
rho3=rho2.subs(a,3);
print('dr1=',(dr1))
print('dr2=',(dr2))
print('rho=',(rho))
print('radius of curvature is',(rho3));
```
