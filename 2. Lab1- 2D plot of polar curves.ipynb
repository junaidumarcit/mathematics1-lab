circle: x=acos(theta), y=bsin(theta), a=3,b=3

```python
import numpy as np
from pylab import *
theta=linspace(0,2*np.pi,200)
x=3*cos(theta)
y=3*sin(theta)
r=sqrt(x**2+y**2)
polar(theta,r,'r--')
show()
```

cardioids: r=a(1+cos(theta)),r=a(1-cos(theta)),a=5

```python
import numpy as np
from pylab import *
theta=linspace(0,2*np.pi,100)
r1=4*(1+cos(theta))
polar(theta,r1,'b--')
r2=-4*(1+cos(theta))
polar(theta,r2,'b--')
show()
```

four leaved rose: r=2(cos(2theta))

```python
import numpy as np
from pylab import *
theta=linspace(0,2*np.pi,100)
r1=2*abs(cos(3*(theta)))
polar(theta,r1,'b--')
show()
```

```python
import numpy as np
from pylab import *
theta=linspace(0,2*np.pi,100)
r1=3+5*theta
polar(theta,r1,'b--')
show()
```
