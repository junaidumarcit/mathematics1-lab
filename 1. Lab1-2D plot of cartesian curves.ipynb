```python
#circle
from sympy import plot_implicit, symbols, Eq
x,y=symbols('x y')
p= plot_implicit(Eq(x**2+y**2,4),(x,-4,4),(y,-4,4),title='circle:$x^2+y^2=4$')
```

```python
#Strophoid
from sympy import plot_implicit,symbols,Eq
x,y=symbols('x y')
p1= plot_implicit(Eq(y**2*(3-x),x**2*(3+x)),(x,-5,5),(y,-5,5), title='strophoid:$y^2(a-x)=x^2(a+x),a=3$')
```

```python
#lemniscate
from sympy import plot_implicit, symbols, Eq
x,y=symbols('x y')
p1=plot_implicit(Eq(y**2*2*(3-x), x**2*(3 - x**2)),(x,-5,5),(y,-5,5),title='lemniscate:$a^2*y^2=x^2(a^2-x^2)$')
```

folium of decats:x^3+y^3=3axy

```python
from sympy import plot_implicit, symbols, Eq
x,y=symbols('x y')
p1=plot_implicit(Eq(x**3+y**3,3*x*y),(x,-5,5),(y,-5,5),title='folium of decats:$x^3+y^3=3axy$')
```

astroid:x^2/3=y^2/3=a^2/3

```python
from sympy import plot_implicit, symbols, Eq
x,y=symbols('x y')
p1=plot_implicit(Eq(x**(2/3)+y**(2/3), 2**(2/3)),(x,-5,5),(y,-5,5),title='astroid:$x^(2/3)+y^(2/3)=a^(2/3)$')
```
